# loading data

```python
try:
    my_data.close()
except:
    pass
my_path = "/reg/data/ana13/sxr/sxri0414/res/hd5/hdf5/"
my_data = h5py.File(my_path+"sxri0414run60.h5",'r')
```

# constructing data matrix for clustering

```python
list(my_data['ebeam'].keys())
```

```python
photon_energy_scattering_product = np.array(my_data['ebeam/photon_energy'])*np.array(my_data['scattered_svd_estimated'])
```

```python
my_relevant_keys = ['GMD/relativeEnergyPerPulse',
                    #'evr/code_162',
                    #'evr/code_141',
                    #'evr/code_142',
                    'delayStage',
                    'gas_detector/f_12_ENRC',
                    'gas_detector/f_22_ENRC',
                    'gas_detector/f_11_ENRC',
                    'gas_detector/f_21_ENRC',
                    'ebeam/L3_energy',
                    'laser_power',
                    'recalculated_atm',                    
                    'recalculated_atm_rms',
                    'scattered_svd_estimated',
                    'ebeam/photon_energy'
                   ]

X = np.array([np.array(my_data[i]) for i in my_relevant_keys])

X = np.vstack([X,photon_energy_scattering_product])

X=X.transpose()
my_mask = (np.array(X[:,4])>0.25)
X = X[my_mask,:]
```

```python
plt.hist(X[:,4],bins=100)
plt.show()
```

# Making sure clustering works on data

```python
row_range = range(X.shape[0])
row_shuffling = np.array(random.sample(row_range,len(row_range))) 
Y = np.nan_to_num(X[row_shuffling][:7000])
Z = 0+Y
```

```python
Y=Y.transpose()
row_sums = (Y.std(axis=1))
Y = Y / row_sums[:, np.newaxis]
Y=Y.transpose()
```

```python
spectral = cluster.SpectralClustering(n_clusters=2,
                                      assign_labels="discretize",
                                      #random_state=0,
                                      gamma=1e-2,
                                      #affinity = 'nearest_neighbors',
                                      #n_neighbors=8
                                      affinity = 'rbf'                                      
                                     )
spectral.fit(np.nan_to_num(Y))
```

```python
plt.scatter(Z[:,-2],Z[:,-0],c=spectral.labels_,alpha = 0.1)
plt.show()
```

```python
to_hist = []
for i in list(set(spectral.labels_)):
    my_mask = (spectral.labels_==i)
    to_hist.append(Z[:,-2][my_mask])

plt.hist(to_hist,bins=100,histtype='step',log=False)
#plt.hist(to_hist,bins=50,histtype='step')

plt.show()
```

```python
plt.scatter(Z[:,0],Z[:,-3],c=spectral.labels_,alpha = 0.1,cmap='jet')
plt.show()
```

```python

plt.semilogy(np.arange(spectral.n_clusters),np.sort([sum(spectral.labels_==i) for i in range(spectral.n_clusters)]),'o')
plt.show()
```

```python
def normalize_matrix(Q):
    Q=Q.transpose()
    row_sums = (Q.std(axis=1))
    Q = Q / row_sums[:, np.newaxis]
    return np.nan_to_num(Q.transpose()),row_sums

Q,normalizations = normalize_matrix(Q)
```

# Getting number of clusters

```python
 #(treating rbf kernel search space as scalar. Could hack by scaling mahalanobis distance along different axes.)
```

```python
row_range = range(Q.shape[0])
row_shuffling = np.array(random.sample(row_range,len(row_range))) 
Z = np.nan_to_num(Q[row_shuffling][:750])
Z.shape
```

```python
temp_list = []
median_map = []
for i in [0.2]:
    my_n_clusters = spectral_cluster_support.get_number_of_clusters(Z,
                                            n_iterations             = 200,
                                            min_n_clusters           = 1,
                                            max_n_clusters           = 20,
                                            regularization_parameter = 1e-14,
                                            affinity                 ='rbf',
                                            gamma                    =i
                                           )
    condition_number,k_range,median_MSE,mean_MSE,std_MSE =  my_n_clusters
    
    plt.semilogy(k_range,median_MSE,'o')
    #plt.ylim(1.9e2,2.2e2)
    plt.show()
    estimated_number_clusters = k_range[np.argmin(median_MSE)]
    print(str(i)+", "+str(estimated_number_clusters))
    temp_list.append([i,estimated_number_clusters])
    median_map.append(median_MSE)
    #if(k_range[np.argmin(median_MSE)]>=4): break
    
```
