```python
base_dict                                      = {}
scan_dict                                      = {}

df                                             = pd.DataFrame()


base_dict['n_iterations']                      = np.array([100],dtype=int)
base_dict['n_samples']                         = np.array([150],dtype=int)
base_dict['affinity']                          = np.array(['rbf'],dtype=str)

base_dict['regularization_parameter']          = 10**np.arange(-11.0,-8,0.3)
base_dict['centers']                           = np.arange(1,13)
base_dict['n_features']                        = np.array([1,2,7,8],dtype=int)
base_dict['cluster_std']                       = np.array([0.2,0.4,0.6,0.8],dtype=float)
base_dict['random_state']                      = np.array([151,170],dtype=int)
base_dict['gamma']                             = np.array([151,170],dtype=int)


#scan_dict['n_features']                        = np.arange(1,8)
#scan_dict['cluster_std']                       = np.arange(0.01,1,0.2)
#scan_dict['random_state']                      = np.array([151,170],dtype=int)


#base_iterator = itertools.product([base_dict[i] for i in base_dict])
def make_frame(base_dict):

    this_df                                         = pd.DataFrame()
    base_iterator                                   = itertools.product(*base_dict.values())

    my_list                                         = np.array(list(base_iterator))
    
    for i,key in enumerate(base_dict): this_df[key] = np.array(my_list[:,i],dtype=base_dict[key].dtype)
    
    return this_df

df = make_frame(base_dict)
```

```python
for i in scan_dict:df = df.append(make_frame({**base_dict,**{i:scan_dict[i]}}))
```

```python
n_samples = 150
random_state = 170
n_centers = 8
regularization_parameter = 1e-9
random_state=151
cluster_std = 0.4
n_features = 2
#X = datasets.make_blobs(n_samples=n_samples, random_state=random_state,centers = n_centers,n_features=n_features,cluster_std=cluster_std)[0]
#my_dict = {"n_samples":n_samples,"random_state":random_state,"centers":n_centers,"cluster_std":cluster_std,"n_features":n_features}
#X = datasets.make_blobs(**my_dict)[0]


sig = inspect.signature(datasets.make_blobs)
result_dict = {}
for j in [{i:df[i].iloc[60]} for i in set(sig.parameters.keys()).intersection(set(df.keys()))]:
    result_dict.update(j)

X = datasets.make_blobs(**result_dict)[0]
plt.scatter(X[:,0],X[:,1],alpha=0.2)
plt.show()
```

```python
def get_number_of_clusters_df(*args,**kwargs):
    df = pd.DataFrame()
    temp_list = spectral_cluster_support.get_number_of_clusters(*args,**kwargs)
    condition_number,df['k_range'],df['median_MSE'],df['mean_MSE'],df['std_MSE'] = temp_list
    df['condition_number'] = condition_number*np.ones(df['k_range'].shape[0])
   
    return df

def get_relevant_keys_from_df(df,target_function):
    target_dict = {}
    sig = inspect.signature(target_function)
    for j in [{i:df[i]} for i in set(sig.parameters.keys()).intersection(set(df.keys()))]:
        target_dict.update(j)
    return target_dict

results_df = pd.DataFrame()
df_index_iterator = iter(range(len(df.index)))
```

```python
for i in df_index_iterator:
        
    #only passing the relevant kwargs to make blobs. 
    make_blob_dict = get_relevant_keys_from_df(df.iloc[i],datasets.make_blobs)
    
    get_n_cluster_dict = get_relevant_keys_from_df(df.iloc[i],spectral_cluster_support.get_number_of_clusters)
    get_n_cluster_dict.update({'min_n_clusters':1,'max_n_clusters':20})
    
    X = datasets.make_blobs(**make_blob_dict)[0]
    
    temp_df = get_number_of_clusters_df(X,**get_n_cluster_dict)
    
    for j in df: temp_df[j] = df[j].iloc[i]
    
    results_df = pd.concat([results_df,temp_df])
    
    if(i%100==1):
         print(str(len(results_df))
         +", "+str(set(results_df['random_state']))
         +", "+str(set(temp_df['random_state']))
         +", "+str(set([df['random_state'].iloc[i]]))
         +", "+str(i))

output_df = results_df.copy()
```

```python
git_hash = subprocess.check_output(["git", "rev-parse","HEAD"]).strip()
#git_hash = subprocess.check_output(["git", "-l"]).strip()

output_df = results_df.copy()
metadata_df = pd.DataFrame()
metadata_df['githash'] = np.array([str(git_hash)[2:-1]])
metadata_df['filename'] = np.array(["systematic_studies_of_spectral_clustering"])
```

```python
my_dict = {}
#regularization_parameter_list = 10.0**np.arange(-19,-4)
regularization_parameter_list = [1e-7,3e-8,7e-8,1.0e-8,7e-9,1.0e-9,3e-9,7e-9,1.0e-10,3e-10,7e-10,7e-11,3e-11,1.0e-11,1e-12]
#n_cluster_list = [2,3,4,5,6,7,8]
n_cluster_list = [1,2,3,4,5,6,7,8]

my_iterator = itertools.product(regularization_parameter_list,n_cluster_list)
n_iterations = 100


for j in n_cluster_list:
    
    my_dict[j,'condition_number'] = []
    my_dict[j,'regularization_parameter'] = []
    my_dict[j,'means'] = []
    my_dict[j,'medians'] = []
    
    
    n_samples = 150
    random_state = 170
    n_centers = j
    X = datasets.make_blobs(n_samples=n_samples, centers = n_centers,n_features=2,cluster_std=0.4,random_state=170)[0]
    #X = datasets.make_blobs(n_samples=n_samples, centers = n_centers,n_features=2,cluster_std=0.1+(np.random.rand(n_centers)**2)/10)[0]
    #X =  datasets.make_moons(n_samples=n_samples, noise=.01)[0]
    plt.scatter(X[:,0],X[:,1],alpha=0.2)
    plt.show()


    for i in regularization_parameter_list:
        
        regularization_parameter = i
        condition_number,k_range,median_MSE_list,mean_MSE_list,std_MSE_list = spectral_cluster_support.get_number_of_clusters(X,
                                                                                    n_iterations=n_iterations,
                                                                                    min_n_clusters=1,
                                                                                    max_n_clusters=20,
                                                                                    regularization_parameter=regularization_parameter,
                                                                                    affinity = 'rbf' #'rbf or nearest_neighbors'
                                                                                    )

        my_dict[j,'condition_number'].append(condition_number)
        my_dict[j,'regularization_parameter'].append(i)
        my_dict[j,'means'].append(mean_MSE_list)
        my_dict[j,'medians'].append(median_MSE_list)

        #make_plots(k_range,median_MSE_list,mean_MSE_list)
```
