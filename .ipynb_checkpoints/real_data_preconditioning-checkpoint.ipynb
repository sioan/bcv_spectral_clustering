```python
len(background_subtracted)
```

```python
plt.plot(background_subtracted[11000])
plt.plot(background_subtracted[10000])
#plt.xlim(200,400)
plt.show()
```

```python
def peakFunction(x,a,x0,offset):
    return a*(x-x0)**2+offset
```

```python
i=1000
filtered_signal = (signal.savgol_filter(background_subtracted[i],25,2,1)  )    #calculates the rolling slope over 25 points
#the filtered_signal is rolling slope over 25 points. instead of convolving with a step function
win_c = np.argmax(abs(filtered_signal))
initial_guess = [1,win_c,filtered_signal[win_c]]
popt,pcov = optimize.curve_fit(peakFunction,np.arange(win_c-4,win_c+5),(filtered_signal[win_c-4:win_c+5]), p0=initial_guess)


plt.plot(filtered_signal)
plt.plot(np.arange(win_c-40,win_c+50),peakFunction(np.arange(win_c-40,win_c+50),*popt))
#plt.ylim(-100,200)
plt.show()
#zooming in
plt.plot(filtered_signal,'.')
window_size = 20
plt.plot(np.arange(win_c-window_size,win_c+window_size),peakFunction(np.arange(win_c-window_size,win_c+window_size),*popt))
plt.xlim(win_c-150,win_c+150)
plt.show()
```

```python
recalculated_atm = []
recalculated_atm_rms = []
recalculated_atm_cov = []
laser_power          = []
def peakFunction(x,a,x0,offset):
    return a*(x-x0)**2+offset

for i in np.arange(len(background_subtracted)):
    
    if(i%2000==1):
        print(i)
    
    filtered_signal = signal.savgol_filter(background_subtracted[i],25,2,1)      #calculates the rolling slope over 25 points
    win_c = np.argmax(abs(filtered_signal))
    initial_guess = [1,win_c,filtered_signal[win_c]]
    try:
        #fitting peak with a parabola
        popt,pcov = optimize.curve_fit(peakFunction,np.arange(win_c-4,win_c+5),(filtered_signal[win_c-4:win_c+5]), p0=initial_guess)


        recalculated_atm.append(popt[1])
        recalculated_atm_cov.append(pcov[1,1])
        recalculated_atm_rms.append(np.sum(background_subtracted[i]**2)**0.5)
        laser_power.append(np.sum(attm_images[i]*1.0))
    except (RuntimeError,ValueError):
        recalculated_atm.append(-999.0)
        recalculated_atm_cov.append(-999.0)
        recalculated_atm_rms.append(-999.0)
        laser_power.append(-999.0)
```

```python
my_data['recalculated_atm_rms']=recalculated_atm_rms
my_data['recalculated_atm']=recalculated_atm
my_data['laser_power']=laser_power
my_data.close()
```
