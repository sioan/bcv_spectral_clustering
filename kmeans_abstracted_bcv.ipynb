# Cross validation for determning number of clusters 
Based on "Estimating the number of clusters using cross-validation by Wei Fu and Patrick O. Perry from the Stern SChool of Business, New York University https://arxiv.org/abs/1702.02658

## basic example from sklearn website (with plotting added)

```python
from sklearn.cluster import KMeans
from sklearn import svm
from sklearn import datasets
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
```

## generating data set

```python
n_samples = 2000
random_state = 170
n_centers = 7
n_features = 8
X = datasets.make_blobs(n_samples=n_samples, centers = n_centers,n_features=n_features,cluster_std=0.1)[0]
```

### embed in higher null feature space 

```python
n_higher_dim = 0+n_centers
X = np.hstack([X,np.zeros([X.shape[0],n_higher_dim])])
X.shape
```

```python
plt.scatter(X[:,0],X[:,1],alpha=0.2)
plt.show()
```

## Making sure clustering algorithm with correct number of centers clusters it correctly

```python
kmeans = KMeans(n_clusters=n_centers, random_state=0).fit(X)
kmeans.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_,alpha = 0.4)
plt.show()
```

## doing the bi-cross validation step

```python
score_list = []
prcnt_rows = 0.5
prcnt_cols = 0.5
cluster_size_array = np.arange(2,15)
n_iterations = 10
for i in cluster_size_array:
    
    score = 0
    
    for k in np.arange(n_iterations):
    
        my_rows = np.arange(X.shape[0])
        my_cols = np.arange(X.shape[1])

        np.random.shuffle(my_rows)
        np.random.shuffle(my_cols)

        my_rows = my_rows
        my_cols = my_cols

        #X_train = X[my_rows,my_cols][:int(my_rows.shape[0]*prcnt_rows),:int(my_cols.shape[0]*prcnt_cols)]
        X_train = X[my_rows][:int(my_rows.shape[0]*prcnt_rows)][:,:int(my_cols.shape[0]*prcnt_cols)]
        y_train = X[my_rows][:int(my_rows.shape[0]*prcnt_rows)][:,int(my_cols.shape[0]*prcnt_cols):]

        X_test = X[my_rows][int(my_rows.shape[0]*prcnt_rows):][:,:int(my_cols.shape[0]*prcnt_cols)]
        y_test = X[my_rows][int(my_rows.shape[0]*prcnt_rows):][:,int(my_cols.shape[0]*prcnt_cols):]


        #step 1: cluster with y_train
        kmeans = KMeans(n_clusters=int(i), random_state=0).fit(y_train)

        #step 2: classify X_train using cluster labels from step 1
        clf = svm.SVC(gamma = 'auto')
        clf.fit(X_train,kmeans.labels_)


        #step 3: predict Apply classifier to X_test
        my_prediction = clf.predict(X_test)

        #step 4: Computer cross validation error
        score = score + np.mean(np.array([np.mean(y_train[kmeans.labels_ == j]) - np.nan_to_num(np.mean(y_test[my_prediction==j])) for j in np.arange(max(kmeans.labels_))])**2)    
   
    score_list.append(score)
    
    
    fig,ax = plt.subplots(1,2)
    fig.set_size_inches(12,6)

    #ax[0].scatter(X_train[:,0],X_train[:,1],c=kmeans.labels_,alpha = 0.4)
    

    ax[0].scatter(X_train[:,0],y_train[:,0],c=kmeans.labels_,alpha = 0.4)
    ax[1].scatter(X_test[:,0],y_test[:,0],c=my_prediction,alpha = 0.4)
    
    
    
    
    plt.show()
```

```python
plt.semilogy(cluster_size_array,score_list)
#plt.ylim(0,10)
plt.show()
```

```python
cluster_size_array[np.argmin(score_list)]
```

## Correcting for strong correlation between clusters as described in section 5 of the paper
## (still in progress)

```python
estimated_n_clusters = cluster_size_array[1+np.argmin(score_list[1:])]
print(estimated_n_clusters)
```

```python
kmeans = KMeans(n_clusters=estimated_n_clusters, random_state=0).fit(X)
kmeans.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_,alpha = 0.4)
plt.show()
```

```python
noise_covariance = np.sum([np.dot((X[kmeans.labels_== i]-np.mean(X[kmeans.labels_== i],axis=0)).transpose(),(X[kmeans.labels_== i]-np.mean(X[kmeans.labels_== i],axis=0))) for i in np.arange(estimated_n_clusters)],axis=0)
```

```python
np.mean(X[kmeans.labels_== 1],axis=0)
```

```python
eig_system = np.linalg.eig(noise_covariance)
```

```python
#rotation_matrix = np.dot(np.dot(eig_system[1],np.diag(eig_system[0]**-0.5)),stats.ortho_group.rvs(eig_system[0].shape[0]))
rotation_matrix = np.dot(eig_system[1],np.diag(eig_system[0]**-0.5))
rotation_matrix = np.dot(rotation_matrix,stats.ortho_group.rvs(eig_system[0].shape[0]))
```

```python
X_rotated = np.dot(X,rotation_matrix)
```

```python
X_test = np.dot(X[kmeans.labels_== 3]-np.mean(X[kmeans.labels_== 3],axis=0),rotation_matrix)+np.mean(X[kmeans.labels_== 3],axis=0)
```

```python
X_test = []
for i in np.arange(estimated_n_clusters):
    X_temp = np.dot(X[kmeans.labels_== i]-np.mean(X[kmeans.labels_== i],axis=0),rotation_matrix)+np.mean(X[kmeans.labels_== i],axis=0)
    X_test.append(X_temp)
    
X_test=np.vstack(X_test)
```

```python
plt.scatter(X_test[:,0],X_test[:,1],alpha = 0.2)
#plt.xlim(2.6,3)
#plt.ylim(2,2.4)
plt.show()
```

```python
fig,ax = plt.subplots(1,2)
fig.set_size_inches(12,6)

ax[0].scatter(X[:,0],X[:,1],c=kmeans.labels_,alpha = 0.2)
ax[0].scatter(X_test[:,0],X_test[:,1],alpha = 0.2)
ax[1].scatter(-X_rotated[:,1],-X_rotated[:,0],c=kmeans.labels_,alpha = 0.2)
plt.show()
```

## turn above into function and apply twice.  Once to reduce spread of each cluster so correlation between clusters is reduced.



```python
score_list = []
prcnt_rows = 0.8
prcnt_cols = 2.0/3.0
cluster_size_array = np.arange(2,15)
n_iterations = 25

def bi_cross_val_score(X,cluster_size,n_iterations):
    
    score = 0
    
    for i in np.arange(n_iterations):
    
            my_rows = np.arange(X.shape[0])
            my_cols = np.arange(X.shape[1])

            np.random.shuffle(my_rows)
            np.random.shuffle(my_cols)

            my_rows = my_rows
            my_cols = my_cols

            #X_train = X[my_rows,my_cols][:int(my_rows.shape[0]*prcnt_rows),:int(my_cols.shape[0]*prcnt_cols)]
            X_train = X[my_rows][:int(my_rows.shape[0]*prcnt_rows)][:,:int(my_cols.shape[0]*prcnt_cols)]
            y_train = X[my_rows][:int(my_rows.shape[0]*prcnt_rows)][:,int(my_cols.shape[0]*prcnt_cols):]

            X_test = X[my_rows][int(my_rows.shape[0]*prcnt_rows):][:,:int(my_cols.shape[0]*prcnt_cols)]
            y_test = X[my_rows][int(my_rows.shape[0]*prcnt_rows):][:,int(my_cols.shape[0]*prcnt_cols):]


            #step 1: cluster with y_train
            kmeans = KMeans(n_clusters=int(cluster_size), random_state=0).fit(y_train)

            #step 2: classify X_train using cluster labels from step 1
            clf = svm.SVC(gamma = 'auto')
            clf.fit(X_train,kmeans.labels_)


            #step 3: predict Apply classifier to X_test
            my_prediction = clf.predict(X_test)

            #step 4: Computer cross validation error
            score += np.mean(np.array([np.mean(y_train[kmeans.labels_ == j]) - np.nan_to_num(np.mean(y_test[my_prediction==j])) for j in np.arange(max(kmeans.labels_))])**2)    

    return score
```

```python
my_scores = []
n_cluster_list = np.arange(2,10,1)
for i in n_cluster_list:
    my_scores.append(bi_cross_val_score(X_rotated,i,12))
```

```python
plt.plot(n_cluster_list,my_scores)
#plt.ylim(0,5)
plt.show()
```
